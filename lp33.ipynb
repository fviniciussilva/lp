Com o Colab, pode aproveitar todo o poder das populares bibliotecas do Python para analisar e visualizar dados. A célula de código abaixo utiliza <strong>numpy</strong> para gerar alguns dados aleatórios e <strong>matplotlib</strong> para os visualizar.
```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import sqlite3
import time
import random
import pandas as pd
import matplotlib.pyplot as plt

# =====================================================================
# 1. FUNÇÕES DE BANCO DE DADOS
# =====================================================================
def preparar_banco_de_dados():
    """Cria uma conexão em memória e a estrutura da tabela."""
    conexao = sqlite3.connect(':memory:')  # Banco de dados temporário em RAM
    cursor = conexao.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS leituras_sensor (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            sensor_id INTEGER,
            temperatura REAL,
            umidade REAL
        )
    ''')
    conexao.commit()
    return conexao

def insercao_sequencial(conexao, dados):
    """Insere dados iterando linha a linha (Abordagem ingênua/Onerosa)."""
    cursor = conexao.cursor()
    for registro in dados:
        cursor.execute(
            'INSERT INTO leituras_sensor (sensor_id, temperatura, umidade) VALUES (?, ?, ?)',
            registro
        )
    conexao.commit()

def insercao_em_lote(conexao, dados):
    """Insere dados em um único lote (Abordagem Otimizada)."""
    cursor = conexao.cursor()
    cursor.executemany(
        'INSERT INTO leituras_sensor (sensor_id, temperatura, umidade) VALUES (?, ?, ?)',
        dados
    )
    conexao.commit()

# =====================================================================
# 2. MOTOR DO EXPERIMENTO
# =====================================================================
def executar_experimento():
    tamanhos_amostra = [10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
    resultados = []

    print("Iniciando Benchmarking do SQLite em Memória...")

    for n in tamanhos_amostra:
        # Geração de Mock Data (Lista de tuplas)
        dados_mock = [(random.randint(1, 100), random.uniform(15.0, 40.0), random.uniform(30.0, 90.0)) for _ in range(n)]

        # ---------------------------------------------
        # Avaliação: Inserção Sequencial
        # ---------------------------------------------
        conn_seq = preparar_banco_de_dados()
        inicio_seq = time.perf_counter()
        insercao_sequencial(conn_seq, dados_mock)
        fim_seq = time.perf_counter()
        tempo_seq = fim_seq - inicio_seq
        conn_seq.close()

        # ---------------------------------------------
        # Avaliação: Inserção em Lote (Bulk)
        # ---------------------------------------------
        conn_lote = preparar_banco_de_dados()
        inicio_lote = time.perf_counter()
        insercao_em_lote(conn_lote, dados_mock)
        fim_lote = time.perf_counter()
        tempo_lote = fim_lote - inicio_lote
        conn_lote.close()

        # Registro das métricas
        resultados.append({
            'Volume de Registros (N)': n,
            'Tempo Sequencial (s)': round(tempo_seq, 4),
            'Tempo em Lote (s)': round(tempo_lote, 4),
            'Redução de Tempo (%)': round(((tempo_seq - tempo_lote) / tempo_seq) * 100, 2)
        })

    return pd.DataFrame(resultados)

# =====================================================================
# 3. EXECUÇÃO E VISUALIZAÇÃO DE RESULTADOS
# =====================================================================
if __name__ == "__main__":
    # Rodar teste e gerar Tabela
    df_resultados = executar_experimento()

    # Exibir Tabela Formatada
    print("\n=== RESULTADOS: SEQUENCIAL VS. LOTE (BULK) ===")
    print(df_resultados.to_string(index=False))
    print("==============================================\n")

    # Gerar Gráfico Analítico
    plt.figure(figsize=(10, 6))
    plt.plot(df_resultados['Volume de Registros (N)'], df_resultados['Tempo Sequencial (s)'],
             marker='o', linestyle='-', color='#d62728', linewidth=2, label='Inserção Sequencial (execute)')
    plt.plot(df_resultados['Volume de Registros (N)'], df_resultados['Tempo em Lote (s)'],
             marker='s', linestyle='-', color='#2ca02c', linewidth=2, label='Inserção em Lote (executemany)')

    # Estilização para o Short Paper
    plt.title('Custo de I/O em Banco de Dados: Inserção Sequencial vs. Em Lote (SQLite)', fontsize=14, weight='bold')
    plt.xlabel('Volume de Transações (Quantidade de Registros)', fontsize=12)
    plt.ylabel('Tempo de Execução (Segundos)', fontsize=12)
    plt.legend(loc='upper left', fontsize=11)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()

    # Salvar para publicação
    nome_arquivo = 'grafico.png'
    plt.savefig(nome_arquivo, dpi=300)
    plt.show()

    print(f"✅ Artefato visual salvo no diretório atual como: '{nome_arquivo}'.")
```
